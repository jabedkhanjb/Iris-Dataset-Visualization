### Libraries

```python
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
```

#### Collecting Dataset from Sklearn

```python
from sklearn.datasets import load_iris

iris = load_iris()

print(f"Iris Dataset feature name: {iris.feature_names}")
print(f"Iris Dataset Target name: {iris.target_names}")
```

```python
# print(iris.data) (next, we're converting this data into csv format)
```

```python
df = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
df
```

```python
# Let's see our target data
iris.target
```

Replacing the target data into csv format with the existing dataframe 

```python
df['species'] = iris.target
df
```

We can see in the target data above that it shows values 0, 1, and 2, which actually specify the dataset elements setosa, virginica, and versicolor.

Now let's replace the number with their respective elements name 

```python
df['species'] = df['species'].replace(0, 'setosa')
df['species'] = df['species'].replace(1, 'virginica')
df['species'] = df['species'].replace(2, 'versicolor')
df
```

#### Save the Dataset

```python
df.to_csv('iris_dataset.csv', index=False)
```

# 2-D Scatter Plot - Visualization

```python
iris_dataset.plot(kind='scatter', x='sepal_length', y='sepal_width', color='green');
plt.show()
```

```python
iris_dataset.plot(kind='scatter', x='sepal_length', y='petal_width', color='blue');
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris_dataset, hue='species', height=6).map(plt.scatter, 'sepal_length', 'sepal_width').add_legend();
```

### Observation

1) By using sepal_length and sepal_width, we can distinguish the setosa flower from others.
2) From the observation of the Seaborn (sns) graph above, we can see the complexity of separating versicolor from virginica due to their overlap.

# 3-D Scatter Plot - Visualization

```python
import plotly.express as px
```

```python
df = px.data.iris()
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_width', color='species')
fig.show()
```

## Pair Plot

```python
sns.set_style('whitegrid')
sns.pairplot(iris_dataset[['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']], hue='species', height=4);
plt.show()
```

### Observations
1. The parameters of petal length and petal width hold significant importance in distinguishing between different types of flowers.
2. Setosa stands out distinctly and can be easily distinguished, whereas there's some overlap between Virnica and Versicolor, making their differentiation more nuanced.
3. By leveraging "lines" and "if-else" conditions, a straightforward model can be devised to classify the various flower types based on these observations.



## Mean , Stardard Deviation & Variance

```python
iris_setosa = iris_dataset[iris_dataset['species']=='setosa']
iris_virginica = iris_dataset[iris_dataset['species']=='virginica']
iris_versicolor = iris_dataset[iris_dataset['species']=='versicolor']
```

```python
print("*********** Mean Values ***********\n")

print(f"Mean values of Iris Setosa {np.mean(iris_setosa['petal_length']):.2f}")
print(f"Mean values of Iris Virginica {np.mean(iris_virginica['petal_length']):.2f}")
print(f"Mean values of Iris Versicolor {np.mean(iris_versicolor['petal_length']):.2f}")
```

## Median, Percentile, Quantile

```python
print("*********** Median Values ***********\n")

print(f"Median values of Iris Setosa {np.median(iris_setosa['petal_length']):.2f}")
print(f"Median values of Iris Virginica {np.median(iris_virginica['petal_length']):.2f}")
print(f"Median values of Iris Versicolor {np.median(iris_versicolor['petal_length']):.2f}")
```

```python
print("*********** Quantiles Values ***********\n")

print(f"Quantiles values of Iris Setosa {np.percentile(iris_setosa['petal_length'], np.arange(0, 100, 25))}")
print(f"Quantiles values of Iris Virginica {np.percentile(iris_virginica['petal_length'], np.arange(0, 100, 25))}")
print(f"Quantiles values of Iris Versicolor {np.percentile(iris_versicolor['petal_length'], np.arange(0, 100, 25))}")
```

```python
print("*********** 90th Percentiles ***********\n")

print(f"90th percentile of Iris Setosa {np.percentile(iris_setosa['petal_length'], 90)}")
print(f"90th percentile of Iris Virginica {np.percentile(iris_virginica['petal_length'], 90)}")
print(f"90th percentile of Iris Versicolor {np.percentile(iris_versicolor['petal_length'], 90)}")
```
